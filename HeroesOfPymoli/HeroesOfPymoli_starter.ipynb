```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(15)
```

## Player Count

* Display the total number of players


```python
total_players = len(pd.unique(pd.Series(purchase_data["SN"])))

total_df=pd.DataFrame({"Total Players": [total_players]}, columns=["Total Players"])
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_unique_items = purchase_data["Item ID"].nunique()
avg_price = purchase_data["Price"].mean()
tot_num_purchases = purchase_data["Purchase ID"].count()
tot_revenue = purchase_data["Price"].sum()

summary_df = pd.DataFrame({"Number of Unique Items": [num_unique_items], 
                           "Average Purchase Price": [avg_price],
                           "Total Number of Purchases": [tot_num_purchases], 
                           "Total Revenue": [tot_revenue]}, 
                  columns=["Number of Unique Items",
                           "Average Purchase Price",
                           "Total Number of Purchases",
                           "Total Revenue"])
summary_df['Average Purchase Price'] = summary_df['Average Purchase Price'].map('${:.2f}'.format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map('${:.2f}'.format)
summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_purchase_data = purchase_data.groupby(["SN"])

purchase_count = pd.DataFrame(grouped_purchase_data["SN"].count())
purchase_count = purchase_count.rename(columns={"SN": "Purchase Count"})
purchase_count=purchase_count.reset_index(drop=False)

avg_purchase_price = pd.DataFrame(grouped_purchase_data["Price"].mean())
avg_purchase_price=avg_purchase_price.reset_index(drop=False)
avg_purchase_price=avg_purchase_price.rename(columns={"Price": "Average Purchase Price"})

total_purchase = pd.DataFrame(grouped_purchase_data["Price"].sum())
total_purchase=total_purchase.reset_index(drop=False)
total_purchase=total_purchase.rename(columns={"Price": "Total Purchase Value"})

top_spenders = pd.merge(purchase_count, avg_purchase_price, on="SN")
top_spenders_df = pd.merge(top_spenders, total_purchase, on="SN")
top_spenders_df = top_spenders_df.set_index('SN')

summary = top_spenders_df.sort_values(['Total Purchase Value'], ascending=False)
summary['Average Purchase Price'] = summary['Average Purchase Price'].map('${:.2f}'.format)
summary['Total Purchase Value'] = summary['Total Purchase Value'].map('${:.2f}'.format)
summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop=purchase_data[["Item ID", "Item Name", "Price"]]
item = most_pop.set_index(['Item Name'])
item = item.sort_values(by="Item ID", ascending=True)
item_df = item.drop_duplicates(["Item ID"])
item_price = list(item_df["Price"])

grouped_most_pop = most_pop.groupby(["Item ID","Item Name"])

most_pop_count = pd.DataFrame(grouped_most_pop["Price"].count())
most_pop_count = most_pop_count.rename(columns={"Price": "Purchase Count"})
most_pop_count=most_pop_count.reset_index(drop=False)
most_pop_count=most_pop_count.set_index("Item ID")

total_purchase_pop = pd.DataFrame(grouped_most_pop["Price"].sum())
total_purchase_pop=total_purchase_pop.rename(columns={"Price": "Total Purchase Value"})
total_purchase_pop=total_purchase_pop.reset_index(drop=False)
purchase_value =list(total_purchase_pop["Total Purchase Value"])

most_pop_count["Item Price"] = item_price
most_pop_count["Total Purchase Value"] = purchase_value
most_pop_count=most_pop_count.reset_index(drop=False)
most_pop_count = most_pop_count.set_index(["Item ID", "Item Name"])
most_pop_count = most_pop_count.sort_values(by="Purchase Count", ascending=False)
most_pop_count['Item Price'] = most_pop_count['Item Price'].map('${:.2f}'.format)
most_pop_count['Total Purchase Value'] = most_pop_count['Total Purchase Value'].map('${:.2f}'.format)
most_pop_count.head()
```
